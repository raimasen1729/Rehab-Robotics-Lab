**VELOCITY OF WRIST ANALYSIS**

```python
left_wrist_patient_x = x_coords[:, 0]
right_wrist_patient_x = x_coords[:, 6]
left_wrist_therapist_x = x_coords[:, 7]
right_wrist_therapist_x = x_coords[:, 13]

left_wrist_patient_y = y_coords[:, 0]
right_wrist_patient_y = y_coords[:, 6]
left_wrist_therapist_y = y_coords[:, 7]
right_wrist_therapist_y = y_coords[:, 13]
```

```python
# Velocity of patient's left wrist

velocity_left_p = []
for i in range(1, len(left_wrist_patient_x)):
  v = np.sqrt((left_wrist_patient_x[i] - left_wrist_patient_x[i-1])**2 + (left_wrist_patient_y[i] - left_wrist_patient_y[i-1])**2)
  velocity_left_p.append(v * 30)
plt.figure(figsize=(25,8))
plt.title("Velocity of left wrist for video8")
plt.xlabel("Frames")
plt.ylabel("Velocity (in pixel/s)")
plt.plot(velocity_left_p[100:3199])
```

```python
# Velocity of patient's right wrist

velocity_right_p = []
for i in range(1, len(left_wrist_patient_x)):
  v = np.sqrt((right_wrist_patient_x[i] - right_wrist_patient_x[i-1])**2 + (right_wrist_patient_y[i] - right_wrist_patient_y[i-1])**2)
  velocity_right_p.append(v * 30)
plt.figure(figsize=(25,8))
plt.title("Velocity of right wrist for video8")
plt.xlabel("Frames")
plt.ylabel("Velocity (in pixel/s)")
plt.plot(velocity_right_p[100:3199])
```

**SHOELACE METHOD**

The shoelace formula or shoelace algorithm is a mathematical algorithm to determine the area of a simple polygon whose vertices are described by their Cartesian coordinates in the plane. The method consists of cross-multiplying corresponding coordinates of the different vertices of a polygon to find its area.

```python
def PolyArea(x,y):
    return 0.5*np.abs(np.dot(x,np.roll(y,1))-np.dot(y,np.roll(x,1)))
```
