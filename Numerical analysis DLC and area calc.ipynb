```python
#first frame area
x_0_vec = x_coords[0,:]
y_0_vec = y_coords[0,:]
```

```python
def PolyArea(x,y):
    return 0.5*np.abs(np.dot(x,np.roll(y,1))-np.dot(y,np.roll(x,1)))
```

```python
print(PolyArea(x_0_vec,y_0_vec))
```

```python
#all frames area
area = []
for i in range(x_coords.shape[0]):
  area.append(PolyArea(x_coords[i,:],y_coords[i,:]))
```

```python
plt.figure(figsize=(25,8))
plt.title("Area over frames for video8")
plt.xlabel("Frames")
plt.ylabel("Area (sq.pixels)")
plt.plot(area[100:5600])
```
